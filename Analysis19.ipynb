Первым делоим ипротируем необходимые библиотеки.

```python
import pandas as pd 
import numpy as np
```

Далее сипользуя команды считываем данные с файла(мой номер 19)

```python
#считывание данных с файла
df = pd.read_csv('vgsales_19.csv')
df
```

Для иследования, узнаем какой размер у наших данных. 

```python
#размеры данных
df.shape 
```

Выведем данные с конца

```python
df.tail()
```

Теперь выведем первые 10 строк данных.

```python
#вывод данных первых 10 строк
df.head(10)
```

Посмотрим полную информацию о датасете, узнаем больше о столбцах, с которыми имеем дело.

```python
#информация о датасете
df.info()
```

Вернемся к данным и посмотрим основные статистические показатели.

```python
#основные статистические показатели
df.describe()
```

Теперь необходимо проверить данные на наличие пропусков.

```python
#проверка наличия пропусков
df.isnull().sum()
```

Также можно вывести размер данных более в удобном формате.

```python
#размер данных
line_count = df.shape[0]
column_count = df.shape[1]
print('Всего строк: {}'.format(line_count))
print('Всего колонок: {}'.format(column_count))
```

Произведем обработку датасета, удалим столбец 'NA_Sales'.

```python
#удаление данных столбца 
df = df.drop(columns=['NA_Sales'])
df
```

Так как в данных есть дубликаты, от них нужно избавиться, удалим их.

```python
#удалени дубликатов(10 дубликатов вышло)
df = df.drop_duplicates()
df.shape
```

Также можно еще раз проверить насколько пусты значения и на основе этих данных сделать вывод.

```python
#обработка пропусков
df.isnull().mean() * 100
```

Больше всего пропусков было в столбце 'Year', таким образом, переменная подлежит удалению.

```python
df = df.drop(columns=['Year'])
df
```

Теперь заменим значения в столбце 'Publisher', чтобы убрать пропуски. Но так же можно было использовать подстановщик, но только к столбцу 'Year'. А так как столбец удален изза множества пропусков, можно просто заменить значения и избавится от пропусков.

```python
df = df.replace({'Publisher' :np.nan}, 'Ужас')
df.isnull().mean() * 100
```

И в завершении проведем сортировку значения на месте в столбце 'JP_Sales' и выведем значения которые наиболее популярны в данном столбце.

```python
#Сортировка значений на месте
df.sort_values(by='JP_Sales', ascending=False, inplace=True)
df.head(8)

```
